## Parte 2 — Funções com parâmetros

Os **parâmetros** são as informações que passamos para a função funcionar. Eles ficam entre os parênteses na definição. Uma função sem return serve para executar uma ação, i.e., ela faz algo, mas não produz um valor que você possa guardar ou usar em outro cálculo. O caso mais comum é quando o objetivo da função é apenas exibir/acessar informações na tela.

```python
# EXEMPLO RESOLVIDO
# Função que calcula e imprime a tensão vertical total em uma camada
# σv = γ × z   (peso específico × profundidade)

def imprimir_tensao_vertical(gamma, profundidade):
    sigma_v = gamma * profundidade
    print(f"Profundidade: {profundidade} m")
    print(f"Peso específico (γ): {gamma} kN/m³")
    print(f"Tensão vertical total (σv): {sigma_v} kPa")

# Chamando com diferentes valores
imprimir_tensao_vertical(18, 5)
print("---")
imprimir_tensao_vertical(20, 3)
```

## Parte 3 — Funções com `return`

Quando a função usa `return`, ela **devolve um valor** que pode ser armazenado em uma variável ou usado diretamente em cálculos.

```python
# EXEMPLO RESOLVIDO
# Função que RETORNA a tensão vertical efetiva
# σ'v = σv - u   (tensão vertical total - poropressão)
# onde u = γw × hw  (peso específico da água × altura da coluna d'água)

def tensao_efetiva(sigma_total, gamma_w, altura_agua):
    u = gamma_w * altura_agua        # poropressão
    sigma_efetiva = sigma_total - u
    return sigma_efetiva

# Usando o valor retornado
sigmaV = 18 * 10   # tensão total a 10m, γ = 18 kN/m³
sigma_ef = tensao_efetiva(sigmaV, 9.81, 6)  # lençol a 4m de profundidade

print(f"Tensão total: {sigmaV} kPa")
print(f"Tensão efetiva: {sigma_ef:.1f} kPa")
```

## Parte 4 — Funções com valor padrão nos parâmetros

É possível definir um **valor padrão** para um parâmetro. Assim, se o usuário não informar esse valor, a função usa o padrão automaticamente.

```python
# EXEMPLO RESOLVIDO
# O peso específico da água (gamma_w) quase sempre é 9.81 kN/m³
# Definimos isso como valor padrão para não precisar informar toda vez

def poropressão(altura_agua, gamma_w=9.81):
    u = gamma_w * altura_agua
    return u

# Usando o valor padrão (sem informar gamma_w)
u1 = poropressão(5)
print(f"Pressão de poros (padrão): {u1:.2f} kPa")

# Sobrescrevendo o valor padrão
u2 = poropressão(5, gamma_w=10.0)
print(f"Pressão de poros (γw=10): {u2:.2f} kPa")
```

## Parte 5 — Combinando funções

Funções podem **chamar outras funções**. Isso permite dividir problemas complexos em partes menores e reutilizáveis.

```python
# EXEMPLO RESOLVIDO
# Calculando tensão efetiva usando duas funções separadas

def tensao_total(gamma, z):
    return gamma * z

def poropressao(gamma_w, hw):
    return gamma_w * hw

def tensao_efetiva(gamma, z, gamma_w, hw):
    sigma_v = tensao_total(gamma, z)       # chama a primeira função
    u = poropressao(gamma_w, hw)         # chama a segunda função
    return sigma_v - u

# Solo: γ = 19 kN/m³, profundidade = 10m
# Lençol a 4m de profundidade → hw = 10 - 4 = 6m
sigma_ef = tensao_efetiva(19, 10, 9.81, 6)
print(f"Tensão efetiva: {sigma_ef:.1f} kPa")
```

```python
# EXERCÍCIO 5B — Desafio
# Perfil de tensões em multicamadas
#
# Crie uma função tensao_multicamada(camadas) que receba uma
# lista de tuplas no formato (gamma, espessura) e calcule
# a tensão vertical acumulada ao longo do perfil.
#
# A função deve imprimir uma tabela como:
#   Camada 1: γ=18 kN/m³, H=3m → σv = 54.0 kPa
#   Camada 2: γ=20 kN/m³, H=2m → σv = 94.0 kPa
#   Camada 3: γ=16 kN/m³, H=5m → σv = 174.0 kPa
#
# Dica: acumule a tensão a cada iteração com um for loop
#
# Teste com:
camadas = [
    (18, 3),   # γ=18 kN/m³, espessura=3m
    (20, 2),   # γ=20 kN/m³, espessura=2m
    (16, 5),   # γ=16 kN/m³, espessura=5m
]

# SEU CÓDIGO AQUI
```

```python
def tensao_multicamada(camadas):
    sigma_acumulada = 0                          # começa do zero (superfície)

    for i, (gamma, espessura) in enumerate(camadas, start=1):   # desempacota cada tupla
        sigma_acumulada += gamma * espessura     # acumula a tensão da camada
        print(f"Camada {i}: γ={gamma} kN/m³, H={espessura}m → σv = {sigma_acumulada:.1f} kPa")

camadas = [
    (18, 3),
    (20, 2),
    (16, 5),
]

tensao_multicamada(camadas)
```
