## 4. Aplicação: Tensão Vertical Total

A **tensão vertical total** num ponto do solo é dada por:

$$\sigma_v = \gamma \times z$$

Onde:
- $\gamma$ = peso específico do solo (kN/m³)
- $z$ = profundidade (m)
- $\sigma_v$ = tensão vertical total (kPa)

```python
gamma = 18.0   # kN/m³
z = 5.0        # m

sigma_v = gamma * z

print("Tensão vertical total:", sigma_v, "kPa")
```

---
## 5. Aplicação: Tensão Efetiva

Considerando o nível d'água, a **tensão efetiva** é:

$$\sigma'_v = \sigma_v - u$$

Onde $u = \gamma_w \times h_w$ é a pressão de poros.

```python
gamma   = 18.0   # Peso específico do solo (kN/m³)
gamma_w = 10.0   # Peso específico da água (kN/m³)
z       = 5.0    # Profundidade total (m)
hw      = 2.0    # Altura da coluna d'água acima do ponto (m)

sigma_v  = gamma * z          # Tensão total
u        = gamma_w * hw       # Pressão de poros
sigma_ef = sigma_v - u        # Tensão efetiva

print("Tensão total:   ", sigma_v,  "kPa")
print("Pressão de poros:", u,       "kPa")
print("Tensão efetiva: ", sigma_ef, "kPa")
```

---
## 6. Input — Pedindo dados ao usuário

Com `input()` o programa pode pedir um valor ao usuário durante a execução. Útil para fazer uma calculadora interativa!

```python
# O usuário digita os valores
gamma = float(input("Digite o peso específico do solo (kN/m³): "))
z     = float(input("Digite a profundidade (m): "))

sigma_v = gamma * z

print("\nTensão vertical total:", sigma_v, "kPa")
```

> 💡 **Atenção:** usamos `float()` para converter o texto digitado em número decimal.
