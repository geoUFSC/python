# PARTE 1 — Indexing and Slicing

```python
nspt = [2, 5, 8, 12, 18, 25, 30]
#      -7 -6 -5  -4  -3  -2  -1

print(nspt[-3:])   # → [18, 25, 30]        últimas 3 camadas
print(nspt[:-2])   # → [2, 5, 8, 12, 18]  tudo menos as 2 últimas
```

## 1.1 Passo (step)

```python
nspt = [2, 5, 8, 12, 18, 25, 30]

print(nspt[::2])   # → [2, 8, 18, 30]               pega de 2 em 2
print(nspt[::-1])  # → [30, 25, 18, 12, 8, 5, 2]    inverte a lista
```

## Aplicação 01: separar camadas pelo nível d'água

```python
nspt = [2, 5, 8, 12, 18, 25, 30]
nivel_agua = 3   # NA está no índice 3

acima_NA  = nspt[:nivel_agua]   # camadas acima do NA
abaixo_NA = nspt[nivel_agua:]   # camadas abaixo do NA

print("Acima do NA: ", acima_NA)
print("Abaixo do NA:", abaixo_NA)
```

## Aplicação 02: últimas leituras de temperatura de solo

```python
# Temperatura do solo (°C) medida ao longo de 10 dias
temperaturas = [18.2, 17.9, 18.5, 19.1, 20.3, 21.0, 20.8, 19.7, 18.9, 18.1]

ultimas_3    = temperaturas[-3:]    # últimas 3 leituras
primeira_semana = temperaturas[:7]  # primeiros 7 dias

print("Últimas 3 leituras:  ", ultimas_3)
print("Primeira semana:     ", primeira_semana)
print("Média últimos 3 dias:", sum(ultimas_3) / len(ultimas_3))
```

# PARTE 2 — Métodos de Lista

Listas têm métodos prontos para operações comuns. Veja os mais úteis:

## 2.1. append, insert, remove, pop

```python
nspt = [2, 5, 8, 12]

nspt.append(18)       # adiciona ao final
print("append:", nspt)

nspt.insert(2, 99)    # insere o valor 99 no índice 2
print("insert:", nspt)

nspt.remove(99)       # remove o valor 99
print("remove:", nspt)

ultimo = nspt.pop()   # remove e retorna o último
print("pop:   ", nspt, "→ removeu:", ultimo)
```

## 2.2. sort, reverse, count, index

```python
nspt = [12, 3, 25, 7, 18, 3, 5]

print("count(3):  ", nspt.count(3))    # quantas vezes o valor 3 aparece
print("index(25): ", nspt.index(25))   # em qual índice está o valor 25

nspt.sort()
print("sort:      ", nspt)             # ordena crescente

nspt.reverse()
print("reverse:   ", nspt)             # inverte a ordem
```

## 2.3. min, max, sum, len

```python
# Precipitação diária (mm) durante uma semana
chuva = [0.0, 5.2, 18.7, 32.1, 12.4, 0.0, 3.5]

print("Mínima: ", min(chuva),        "mm")
print("Máxima: ", max(chuva),        "mm")
print("Total:  ", sum(chuva),        "mm")
print("Dias:   ", len(chuva),        "dias")
print("Média:  ", sum(chuva)/len(chuva), "mm/dia")
```

## 2.4. Aplicação: análise de série de NSPT

```python
nspt = [3, 7, 5, 12, 18, 9, 25, 30, 4, 22]

print("Número de camadas:", len(nspt))
print("NSPT mínimo:      ", min(nspt))
print("NSPT máximo:      ", max(nspt))
print("NSPT médio:       ", sum(nspt) / len(nspt))

nspt_ordenado = sorted(nspt)   # sorted() retorna nova lista sem alterar a original
print("Ordenado:         ", nspt_ordenado)
```

# PARTE 3 — Métodos de String

Strings também têm métodos prontos. Muito úteis para processar nomes, classificações e dados lidos de arquivos.

## 3.1. upper, lower, strip, replace

```python
solo = "  Argila Siltosa  "

print(solo.upper())     # → "  ARGILA SILTOSA  "   tudo maiúsculo (interessante para padronização)
print(solo.lower())     # → "  argila siltosa  "   tudo minúsculo (interessante para padronização)
print(solo.strip())     # → "Argila Siltosa"        remove espaços das bordas (interessante para limpar bd)
print(solo.replace("Siltosa", "Arenosa"))   # substitui trecho
```

## 3.2. find, startswith, endswith, in

```python
descricao = "Argila siltosa de alta plasticidade"

print(descricao.find("siltosa"))          # → 7    posição onde começa
print(descricao.find("areia"))            # → -1   não encontrou
print(descricao.startswith("Argila"))     # → True começa com "Argila"?
print(descricao.endswith("plasticidade")) # → True termina com "plasticidade"?
print("siltosa" in descricao)             # → True contém "siltosa"?
```

## 3.3. split e join

```python
# split — quebra string em lista
linha = "SP-01, SP-02, SP-03, SP-04"
furos = linha.split(", ")
print("split:", furos)     # → ['SP-01', 'SP-02', 'SP-03', 'SP-04']

# join — une lista em string
unido = " | ".join(furos)
print("join: ", unido)     # → SP-01 | SP-02 | SP-03 | SP-04
```

## Aplicação 01: lendo uma linha de dados de sondagem

```python
# Simula uma linha lida de um arquivo de texto
linha = "SP-01;12.5;Argila siltosa;18"
print(linha)
partes = linha.split(";")
print(partes)
print(type(partes))

furo        = partes[0]
profundidade = float(partes[1])
solo        = partes[2]
nspt        = int(partes[3])

print(f"Furo:         {furo}")
print(f"Profundidade: {profundidade} m")
print(f"Solo:         {solo}")
print(f"NSPT:         {nspt}")
```

> 💡 Esse padrão — ler uma linha, fazer split e extrair os valores — é exatamente o que acontece quando você lê arquivos `.csv` em Python.

## Aplicação 02: filtrar furos pelo nome

```python
furos = ["SP-01", "SM-01", "SP-02", "SM-02", "SP-03"] #SM: Sondagem Mista (percussão + rotativa)

# Filtrar apenas furos do tipo SP
furos_sp = [i for i in furos if i.startswith("SP")] #list comprehension:  É uma forma compacta de escrever um loop for
print("Furos SP:", furos_sp)

'''#ou com loop for tradicional
furos_sp = []
for i in furos:
  if i.startswith("SP"):
    furos_sp.append(i) '''

# Filtrar apenas furos do tipo SM
furos_sm = [i for i in furos if i.startswith("SM")]
print("Furos SM:", furos_sm)
```

## Exercício 1 — Slicing de precipitação

Dada a lista de precipitação mensal (mm) abaixo, use slicing para:
- Extrair o primeiro trimestre (Jan–Mar)
- Extrair o segundo semestre (Jul–Dez)
- Calcular a média do segundo semestre

```python
# Precipitação mensal (mm) — Jan a Dez
precipitacao = [180, 160, 140, 90, 60, 40, 35, 50, 80, 120, 150, 175]

primeiro_trimestre = precipitacao[?:?]
segundo_semestre   = precipitacao[?:]
media_2sem         = sum(segundo_semestre) / len(segundo_semestre)

print("Primeiro trimestre:", primeiro_trimestre)
print("Segundo semestre:  ", segundo_semestre)
print(f"Média 2° semestre:  {media_2sem:.1f} mm")

# Respostas esperadas:
# Primeiro trimestre: [180, 160, 140]
# Segundo semestre:   [35, 50, 80, 120, 150, 175]
# Média 2° semestre:  101.7 mm
```

## 🎯 Resumo do que você aprendeu

| Conceito | O que faz | Exemplo |
|---|---|---|
| `lista[0]` | Indexing — acessa um elemento | Primeira camada da sondagem |
| `lista[-1]` | Indexing negativo — último elemento | Camada mais profunda |
| `lista[1:4]` | Slicing — pega pedaço da lista | Camadas de 1 a 3 |
| `lista[-3:]` | Slicing negativo — últimos N elementos | Últimas 3 camadas |
| `lista[::2]` | Slicing com passo | De 2 em 2 metros |
| `append / pop` | Adiciona / remove da lista | Novo furo / remover leitura |
| `sort / sorted` | Ordena a lista | NSPT em ordem crescente |
| `min / max / sum` | Estatísticas básicas | Análise de precipitação |
| `split / join` | Quebra / une strings | Ler linha de arquivo |
| `find / in` | Busca em string | Filtrar tipo de furo |
| `strip / replace` | Limpa / substitui texto | Padronizar nomes |
