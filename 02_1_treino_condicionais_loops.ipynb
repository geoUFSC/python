## Parte 1 — Condicionais (`if / elif / else`)

Condicionais permitem que o programa tome **decisões** com base em valores. Em geotecnia, usamos isso o tempo todo para classificar solos, verificar limites de segurança e interpretar ensaios.

````python
# EXEMPLO RESOLVIDO
# Classificação de areia ou silte arenoso pelo SPT (ABNT 6484)
#   N <= 4        → "Fofa"
#   5  <= N <= 8  → "Pouco compacta"
#   9 <= N <= 18  → "Medianamente compacta"
#   19 <= N <= 40 → "Compacta"
#   N > 40        → "Muito compacta"

N = 12

if N <= 4:
    classificacao = "Fofa"
elif N <= 8:
    classificacao = "Pouco compacta"
elif N <= 18:
    classificacao = "Medianamente compacta"
elif N <= 40:
    classificacao = "Compacta"
else:
    classificacao = "Muito compacta"

print(f"N = {N} → {classificacao}")
````

## Parte 2 — Loop `for`

Use o `for` quando você sabe **quantas vezes** o loop vai repetir, por exemplo, para percorrer uma lista de amostras ou camadas de solo.

````python
# EXEMPLO RESOLVIDO
# Classificar uma lista de valores N (areia) e imprimir cada resultado

golpes_areia = [2, 6, 14, 25, 55]

for N in golpes_areia:
    if N <= 4:
        classe = "Fofa"
    elif N <= 8:
        classe = "Pouco compacta"
    elif N <= 18:
        classe = "Medianamente compacta"
    elif N <= 40:
        classe = "Compacta"
    else:
        classe = "Muito compacta"

    print(f"N = {N:>3} → {classe}") # :>3 significa: alinhe à direita em um espaço de 3 caracteres
````

## Parte 3 — Loop `while`

Use o `while` quando você **não sabe quantas repetições** serão necessárias — o loop continua até uma condição ser atingida. Em geotecnia, isso aparece em processos como avanço de sondagem ou monitoramento até um limite.

````python
# Simular avanço de sondagem — parada quando N >= 45 (impenetrável)

N = 0
profundidade = 0

while N < 45:
    profundidade += 1
    N = int(input(f"Digite o N do SPT na profundidade {profundidade} m: "))
    print(f"  N = {N} → sondagem continua...")

print(f"\nImpenetrável (N={N}) atingido em {profundidade} m. Sondagem encerrada.")
````

````python
# Simular avanço de sondagem — parada quando N >= 45 (impenetrável)

import random
random.seed(10)  # semente fixa para o exemplo ser reproduzível
profundidade = 0

while True: # condição que jamais se torna False
    profundidade += 1
    N = random.randint(1, 60)   # gera um número inteiro aleatório para N entre 1 e 60, incluindo os dois extremos
    print(f"Prof. {profundidade} m → N = {N}")

    if N >= 45:
        print(f"\nImpenetrável atingido em {profundidade} m. Sondagem encerrada.")
        break


'''No while True a única saída é o break — por isso ele é usado quando você
não consegue escrever a condição de parada na linha do while, mas sabe que
la vai aparecer dentro do loop. No caso da sondagem: você não sabe em qual
metro vai parar, mas sabe que quando N >= 45 aparecer, o break vai encerrar tudo'. ''
````

## Parte 5 — `enumerate`: índice + valor ao mesmo tempo

Quando usamos um `for` normal, o Python entrega apenas o **valor** de cada elemento.
O `enumerate` entrega também a **posição** (índice) — sem precisar criar um contador manual.

```python
# sem enumerate → você controla o contador
i = 1
for valor in lista:
    print(i, valor)
    i += 1

# com enumerate → mais simples
for i, valor in enumerate(lista, 1):   # 1 = começa a contar do 1
    print(i, valor)
```

````python
# EXEMPLO RESOLVIDO
# Listar amostras de solo com número de ordem

amostras = ["Argila mole", "Areia fina", "Silte argiloso", "Areia grossa"]

for i, amostra in enumerate(amostras, 1):
    print(f"Amostra {i}: {amostra}")
````
