# 🔢 PARTE 1 — NumPy

**NumPy** (Numerical Python) é a base de quase toda ciência de dados em Python. Ele fornece **arrays n-dimensionais** muito mais rápidos que listas normais, além de dezenas de funções matemáticas prontas.

### Por que não usar listas Python comuns?

- Listas são flexíveis (aceitam diferentes tipos de dados), mas **lentas** para cálculos numéricos
- Arrays NumPy armazenam dados em blocos contíguos de memória (como C)
- Resultado: até **50× mais rápidos** para operações numéricas

```python
# A convenção universal é importar numpy como 'np'
import numpy as np

print('NumPy versão:', np.__version__)
print('✅ NumPy importado com sucesso!')
```

### 1.1 Criando arrays

```python
# Criar array a partir de uma lista Python
temperaturas = np.array([28.5, 31.0, 29.8, 33.2, 27.1])
print('Array de temperaturas:', temperaturas)
print('Tipo dos dados:', temperaturas.dtype)   # float64
print('Formato (shape):', temperaturas.shape)  # (5,) → 5 elementos
print('Número de elementos:', temperaturas.size)
```

```python
# Arrays especiais — muito úteis no dia a dia

zeros    = np.zeros(5)            # [0. 0. 0. 0. 0.]
uns      = np.ones((3, 3))        # matriz 3×3 toda de 1s
sequencia = np.arange(0, 10, 2)  # [0 2 4 6 8] — como range(), mas NumPy
espacado  = np.linspace(0, 1, 5) # 5 pontos igualmente espaçados entre 0 e 1
aleatorio = np.random.rand(4)    # 4 números aleatórios entre 0 e 1

print('Zeros:    ', zeros)
print('Sequência:', sequencia)
print('Espaçado: ', espacado)
print('Aleatório:', aleatorio.round(3))
print('\nMatriz de 1s (3x3):')
print(uns)
```

### 1.2 Operações vetorizadas

A grande magia do NumPy: **operações aplicadas em todos os elementos de uma vez**, sem precisar de loops `for`.

```python
celsius = np.array([28.5, 31.0, 29.8, 33.2, 27.1])

# Com lista Python (verboso e lento):
fahrenheit_lista = [t * 9/5 + 32 for t in [28.5, 31.0, 29.8, 33.2, 27.1]]

# Com NumPy (rápido e elegante!):
fahrenheit_np = celsius * 9/5 + 32  # opera em todos de uma vez

print('Celsius:   ', celsius)
print('Fahrenheit:', fahrenheit_np.round(1))

# Estatísticas básicas
print(f'\nMédia:      {np.mean(celsius):.1f} °C')
print(f'Máximo:     {np.max(celsius):.1f} °C')
print(f'Mínimo:     {np.min(celsius):.1f} °C')
print(f'Desvio pad: {np.std(celsius):.2f} °C')
print(f'Soma total: {np.sum(celsius):.1f} °C')
```

```python
# Filtragem booleana — muito poderosa!
precipitacao = np.array([0.0, 5.2, 18.7, 0.0, 3.1, 22.4, 0.0])

# Dias em que choveu
choveu = precipitacao[precipitacao > 0]
print(precipitacao > 0) #booleano
print('Dias com chuva (mm):', choveu)

# Onde choveu? (índices)
indices = np.where(precipitacao > 0) #np.where() retorna uma tupla: array([1, 2, 4, 5]), )
print('Índices dos dias com chuva:', indices[0]) #foi projetado para funcionar também em matrizes 2D retornando tupla de (linhas, colunas)

# Dia de maior precipitação
dia_max = np.argmax(precipitacao)
print(f'Maior chuva no dia {dia_max}: {precipitacao[dia_max]} mm')

# Total e média apenas dos dias com chuva
print(f'Total de chuva: {precipitacao.sum():.1f} mm')
print(f'Média nos dias chuvosos: {choveu.mean():.1f} mm')
```

### 1.3 Matrizes (arrays 2D)

Arrays 2D funcionam como tabelas: **linhas × colunas**.

```python
# Exemplo: 3 dias × 3 estações de monitoramento
dados = np.array([
    [28.1, 29.5, 27.8],   # Dia 1 — Estações A, B, C
    [31.0, 32.1, 30.5],   # Dia 2
    [26.3, 27.0, 25.9],   # Dia 3
])

print('Matriz de dados:')
print(dados)
print(f'\nFormato: {dados.shape}  →  {dados.shape[0]} dias, {dados.shape[1]} estações')

# Acessar elementos: [linha, coluna] — índice começa em 0!
print(f'\nDia 1, Estação B: {dados[0, 1]}')  # linha 0, coluna 1
print(f'Dia 2 completo:   {dados[1, :]}')    # linha 1, todas as colunas
print(f'Estação A (todos os dias): {dados[:, 0]}')  # todas linhas, coluna 0

# Estatísticas por eixo
print(f'\nMédia por estação (axis=0): {np.mean(dados, axis=0)}')  # percorre as linhas (colapsa as linhas, sobra uma coluna), resultado por coluna
print(f'Média por dia     (axis=1): {np.mean(dados, axis=1)}')    # resultado por linha
''' mesmo que
media_estacao_A = np.mean(dados[:, 0])  # todas linhas, coluna 0
media_estacao_B = np.mean(dados[:, 1])  # todas linhas, coluna 1
media_estacao_C = np.mean(dados[:, 2])  # todas linhas, coluna 2
media_dia1 = np.mean(dados[0, :])  # linha 0, todas colunas
media_dia2 = np.mean(dados[1, :])
media_dia3 = np.mean(dados[2, :])

Mas o axis é a forma preferida porque é mais curta e mais rápida. Os outros
jeitos ficam trabalhosos quando a matriz é grande — imagina fazer isso com 365
dias e 20 estações.'''
```

# 🐼 PARTE 2 — Pandas

**Pandas** traz o conceito de **DataFrame** — uma tabela com rótulos nas linhas e colunas, como uma planilha Excel dentro do Python.

Duas estruturas principais:
- **Series** → uma coluna (array 1D com índice)
- **DataFrame** → tabela completa (coleção de Series)

```python
import pandas as pd

print('Pandas versão:', pd.__version__)
print('✅ Pandas importado!')
```

### 2.1 Criando um DataFrame

```python
# Criar DataFrame a partir de um dicionário
# Cada chave = nome da coluna, cada valor = lista com os dados

dados = {
    'data':        ['2024-01-15', '2024-01-15', '2024-01-16', '2024-01-16', '2024-01-17', '2024-01-17'],
    'local':       ['Norte', 'Sul', 'Norte', 'Sul', 'Norte', 'Sul'],
    'temperatura': [27.3, 28.1, 25.5, 26.0, 31.2, 30.5],
    'umidade':     [82, 79, 88, 85, 65, 68],
    'chuva_mm':    [0.0, 0.0, 12.3, 8.7, 0.0, 0.0],
}

df = pd.DataFrame(dados)

# Converter coluna de data para o tipo correto
df['data'] = pd.to_datetime(df['data'])

# Exibir a tabela (no Colab aparece formatada automaticamente)
df
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=df)
```

```python
# Inspecionar o DataFrame
print('=== FORMATO ===')
print(df.shape)
print(f'{df.shape[0]} linhas × {df.shape[1]} colunas')

print('\n=== PRIMEIRAS LINHAS ===')
print(df.head(3))  # ou tail(3) para as últimas

print('\n=== TIPOS DE DADOS ===')
print(df.dtypes)

print('\n=== ESTATÍSTICAS ===')
print(df.describe().round(2))
```

### 2.2 Seleção e filtragem

```python
# Selecionar uma coluna → retorna uma Series
print('Coluna temperatura:')
print(df['temperatura'])

print('\nMédia da temperatura:', df['temperatura'].mean().round(1))
```

```python
# Filtrar linhas com condição
dias_com_chuva = df[df['chuva_mm'] > 0]
print('--Dias com chuva:')
print(dias_com_chuva)

print(df['chuva_mm'] > 0) #booleano

# Múltiplas condições: & (AND), | (OR)
quente_e_seco = df[(df['temperatura'] > 29) & (df['umidade'] < 70)]
print('\n','--Dias quentes E secos (temp > 29 E umidade < 70):')
print(quente_e_seco)
print()
print((df['temperatura'] > 29) & (df['umidade'] < 70))
print()

# Filtrar por texto
estacao_norte = df[df['local'] == 'Norte']
print('--Apenas estação Norte:')
print(estacao_norte)
```

```python
# Acessar por posição: .iloc[linha] ou .iloc[linha, coluna]
print('Primeira linha (iloc[0]):')
print(df.iloc[0])

print('\nUltima linha (iloc[-1]):')
print(df.iloc[-1])       # última linha

print('\nLinhas 2 a 4 (iloc[1:4]):')
print(df.iloc[1:4])
```

### 2.3 Criando novas colunas e aplicando funções

```python
# Criar coluna calculada (operação vetorial)
df['temp_fahrenheit'] = df['temperatura'] * 9/5 + 32

# Criar coluna com função personalizada via .apply()
def categoria_chuva(mm):
    if mm == 0:
        return 'sem chuva'
    elif mm < 5:
        return 'fraca'
    elif mm < 20:
        return 'moderada'
    else:
        return 'forte'

df['categoria_chuva'] = df['chuva_mm'].apply(categoria_chuva)

# Mostrar colunas relevantes
df[['data', 'local', 'temperatura', 'temp_fahrenheit', 'chuva_mm', 'categoria_chuva']]
```

### 2.4 Agrupamento com groupby()

`groupby()` é equivalente ao `GROUP BY` do SQL — agrupa linhas por uma coluna e aplica uma função a cada grupo.

```python
# Temperatura média por local
print('Temperatura média por local:')
print(df.groupby('local')['temperatura'].mean().round(1))

print()

# Múltiplas estatísticas de uma vez com .agg()
resumo = df.groupby('local').agg(
    temp_media=('temperatura', 'mean'),
    temp_max=('temperatura', 'max'),
    temp_min=('temperatura', 'min'),
    chuva_total=('chuva_mm', 'sum'),
    n_medicoes=('temperatura', 'count')
).round(1)

print('Resumo completo por local:')
resumo
```

# 📊 PARTE 3 — Matplotlib

**Matplotlib** é a principal biblioteca de visualização do Python. Com ela você cria gráficos de linha, barras, histogramas, dispersão e muito mais.

A convenção é importar `pyplot` como `plt`.

```python
import matplotlib.pyplot as plt

# No Colab, gráficos aparecem diretamente na célula (já é o padrão)
print('✅ Matplotlib importado!')
```

### 3.1 Gráfico de linha

```python
plt.plot?
```

```python
dias = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sáb', 'Dom']
temp = [28.5, 31.2, 26.8, 29.1, 33.5, 27.3, 25.0]

plt.figure(figsize=(10, 5))  # largura × altura em polegadas

plt.plot(dias, temp,
         color='steelblue',
         linewidth=2,
         marker='o',       # marcador em cada ponto
         markersize=7,
         label='Temperatura °C')

# Linha de referência (média)
media = sum(temp) / len(temp) #np.mean(temp)
plt.axhline(media, color='red', linestyle='--', alpha=0.6, label=f'Média ({media:.1f}°C)')

plt.title('Temperatura Semanal', fontsize=14, fontweight='bold')
plt.xlabel('Dia da Semana')
plt.ylabel('Temperatura (°C)')
plt.ylim(20, 38)              # limites do eixo Y
plt.grid(axis='y', alpha=0.3) # grade horizontal suave
plt.legend()
plt.tight_layout()
plt.savefig('temperatura_semanal.png', dpi=150)  # salvar como imagem
plt.show()
print('Gráfico salvo como temperatura_semanal.png')
```

### 3.2 Gráfico de barras

```python
especies  = ['Espécie A', 'Espécie B', 'Espécie C', 'Espécie D']
contagens = [42, 85, 31, 67]
cores     = ['#4CAF50', '#2196F3', '#FF9800', '#9C27B0']

plt.figure(figsize=(8, 5))

barras = plt.bar(especies, contagens, color=cores, edgecolor='white', linewidth=0.8)

# Adicionar valores em cima das barras
for barra, valor in zip(barras, contagens):
    plt.text(barra.get_x() + barra.get_width()/2, barra.get_height() + 1,
             str(valor), ha='center', va='bottom', fontweight='bold')

plt.title('Contagem de Indivíduos por Espécie', fontsize=13)
plt.ylabel('Número de indivíduos')
plt.ylim(0, 100)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

### 3.3 Gráfico de dispersão (scatter)

```python
# Gerar dados simulados de temperatura × umidade
np.random.seed(42)
temp_sim = np.random.normal(28, 3, 60)
umid_sim = 120 - temp_sim * 1.5 + np.random.normal(0, 5, 60)  # correlação negativa

plt.figure(figsize=(7, 5))

plt.scatter(temp_sim, umid_sim,
            alpha=0.6,
            color='coral',
            edgecolors='darkred',
            linewidths=0.5,
            s=60)  # tamanho dos pontos

plt.xlabel('Temperatura (°C)')
plt.ylabel('Umidade (%)')
plt.title('Relação Temperatura × Umidade', fontsize=13)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

# 📂 PARTE 4 — Leitura de Arquivos

Na vida real, os dados chegam em arquivos. Saber ler `.txt` e `.csv` é fundamental para quem trabalha com dados de campo.

### Criando arquivos de exemplo (para usar no Colab)

```python
# No Colab, vamos criar os arquivos de exemplo diretamente pelo Python

# Criar dados.txt
conteudo_txt = """# Dados de monitoramento — Estação Norte
# Formato: data hora temperatura umidade
2024-01-15 08:00 27.3 82
2024-01-15 12:00 32.1 65
2024-01-15 18:00 29.8 71
2024-01-16 08:00 25.5 88
2024-01-16 12:00 30.2 69
2024-01-16 18:00 28.9 74
"""

with open('dados.txt', 'w', encoding='utf-8') as f:
    f.write(conteudo_txt)

print('✅ dados.txt criado!')
print('Conteúdo:')
print(conteudo_txt)
```

```python
# Criar medicoes.csv
conteudo_csv = """data,local,temperatura,umidade,chuva_mm
2024-01-15,Norte,27.3,82,0.0
2024-01-15,Sul,28.1,79,0.0
2024-01-16,Norte,25.5,88,12.3
2024-01-16,Sul,26.0,85,8.7
2024-01-17,Norte,31.2,65,0.0
2024-01-17,Sul,30.5,68,0.0
2024-01-18,Norte,29.8,70,4.2
2024-01-18,Sul,28.9,73,2.1
2024-01-19,Norte,33.1,58,0.0
2024-01-19,Sul,32.4,61,0.0
"""

with open('medicoes.csv', 'w', encoding='utf-8') as f:
    f.write(conteudo_csv)

print('✅ medicoes.csv criado!')
```

### 4.1 Lendo arquivos .txt

```python
# Forma 1: ler o arquivo inteiro de uma vez
with open('dados.txt', 'r', encoding='utf-8') as f:
    tudo = f.read()

print('=== CONTEÚDO COMPLETO ===')
print(tudo)
```

```python
# Forma 2: linha por linha — útil para arquivos grandes e para filtrar

temperaturas = []
umidades = []
horarios = []

with open('dados.txt', 'r', encoding='utf-8') as f:
    for linha in f:
        linha = linha.strip()               # remove espaços e \n
        if linha.startswith('#'):           # pular linhas de comentário
            continue
        if not linha:                       # pular linhas vazias
            continue

        partes = linha.split()              # divide por espaços em branco
        # partes = ['2024-01-15', '08:00', '27.3', '82']

        horarios.append(partes[0] + ' ' + partes[1])
        temperaturas.append(float(partes[2]))
        umidades.append(int(partes[3]))

print('Horários lidos:     ', horarios)
print('Temperaturas (°C):  ', temperaturas)
print('Umidades (%):       ', umidades)
print(f'\nMédia temperatura:  {sum(temperaturas)/len(temperaturas):.1f} °C')
print(f'Máximo temperatura: {max(temperaturas):.1f} °C')
```

# 🚀 PROJETO FINAL — Relatório Completo de Monitoramento

Agora vamos unir tudo: **leitura de arquivo → limpeza → análise → visualização → exportação**.

Este é o ciclo real de qualquer projeto de análise de dados.

```python
# ══════════════════════════════════════════════
# ETAPA 1 — LEITURA E LIMPEZA
# ══════════════════════════════════════════════

df = pd.read_csv('medicoes.csv')
df['data'] = pd.to_datetime(df['data'])

print(f'✅ {len(df)} registros carregados')
print(f'   Valores nulos: {df.isnull().sum().sum()}')

df = df.dropna()  # remover linhas incompletas
print(f'   Após limpeza: {len(df)} registros')
```

```python
# ══════════════════════════════════════════════
# ETAPA 2 — ANÁLISE COM NUMPY E PANDAS
# ══════════════════════════════════════════════

temps = df['temperatura'].values  # extrai como array NumPy

print('📊 ANÁLISE GERAL DE TEMPERATURA')
print(f'   Média:       {np.mean(temps):.1f} °C')
print(f'   Desvio pad.: {np.std(temps):.2f} °C')
print(f'   Máximo:      {np.max(temps):.1f} °C')
print(f'   Mínimo:      {np.min(temps):.1f} °C')
print(f'   Dias > 30°C: {int(np.sum(temps > 30))}')

print()
print('🌧️ ANÁLISE DE CHUVA')
print(f'   Total geral: {df["chuva_mm"].sum():.1f} mm')
print(f'   Dias com chuva: {len(df[df["chuva_mm"] > 0])}')
print(f'   Maior evento: {df["chuva_mm"].max():.1f} mm')

print()
print('📍 RESUMO POR LOCAL')
resumo = df.groupby('local').agg(
    temp_media=('temperatura', 'mean'),
    temp_max=('temperatura', 'max'),
    umidade_media=('umidade', 'mean'),
    chuva_total=('chuva_mm', 'sum'),
).round(1)
print(resumo)
```

```python
# ══════════════════════════════════════════════
# ETAPA 3 — VISUALIZAÇÃO (4 gráficos)
# ══════════════════════════════════════════════

fig, axes = plt.subplots(2, 2, figsize=(13, 9))
fig.suptitle('Relatório de Monitoramento Ambiental', fontsize=15, fontweight='bold', y=1.01)

cores_locais = {'Norte': '#2196F3', 'Sul': '#FF5722'}

# ── Gráfico 1: Temperatura ao longo do tempo (por local)
for local, grupo in df.groupby('local'):
    axes[0,0].plot(grupo['data'], grupo['temperatura'],
                   marker='o', label=local, color=cores_locais[local], linewidth=2)
axes[0,0].set_title('Temperatura ao Longo do Tempo')
axes[0,0].set_ylabel('°C')
axes[0,0].legend()
axes[0,0].grid(alpha=0.3)
axes[0,0].tick_params(axis='x', rotation=30)

# ── Gráfico 2: Chuva acumulada por local (barras)
chuva_local = df.groupby('local')['chuva_mm'].sum()
axes[0,1].bar(chuva_local.index, chuva_local.values,
              color=[cores_locais[l] for l in chuva_local.index],
              edgecolor='white', linewidth=0.8)
for i, (local, val) in enumerate(chuva_local.items()):
    axes[0,1].text(i, val + 0.3, f'{val:.1f} mm', ha='center', fontweight='bold')
axes[0,1].set_title('Chuva Acumulada por Local (mm)')
axes[0,1].set_ylabel('mm')
axes[0,1].grid(axis='y', alpha=0.3)

# ── Gráfico 3: Histograma de temperatura
axes[1,0].hist(temps, bins=8, color='mediumseagreen', edgecolor='white', linewidth=0.8)
axes[1,0].axvline(np.mean(temps), color='red', linestyle='--', label=f'Média: {np.mean(temps):.1f}°C')
axes[1,0].set_title('Distribuição de Temperatura')
axes[1,0].set_xlabel('°C')
axes[1,0].set_ylabel('Frequência')
axes[1,0].legend()
axes[1,0].grid(alpha=0.3)

# ── Gráfico 4: Temperatura × Umidade (scatter por local)
for local, grupo in df.groupby('local'):
    axes[1,1].scatter(grupo['temperatura'], grupo['umidade'],
                      alpha=0.7, label=local, color=cores_locais[local],
                      s=70, edgecolors='white', linewidths=0.5)
axes[1,1].set_xlabel('Temperatura (°C)')
axes[1,1].set_ylabel('Umidade (%)')
axes[1,1].set_title('Temperatura × Umidade')
axes[1,1].legend()
axes[1,1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('relatorio_completo.png', dpi=150, bbox_inches='tight')
plt.show()
print('\n✅ Relatório salvo: relatorio_completo.png')
```

```python
# ══════════════════════════════════════════════
# ETAPA 4 — EXPORTAR DADOS PROCESSADOS
# ══════════════════════════════════════════════

# Enriquecemos o DataFrame com colunas calculadas
df['temp_fahrenheit'] = (df['temperatura'] * 9/5 + 32).round(1)
df['alerta_calor']    = df['temperatura'].apply(lambda t: 'SIM' if t > 30 else 'não')
df['mes']             = df['data'].dt.month
df['dia_semana']      = df['data'].dt.day_name(locale='pt_BR.utf-8') if False else df['data'].dt.day_name()

# Salvar CSV final
df.to_csv('dados_processados.csv', index=False)
print('✅ dados_processados.csv salvo!')

print()
print('Prévia do arquivo final:')
df[['data', 'local', 'temperatura', 'temp_fahrenheit', 'chuva_mm', 'alerta_calor']].head(6)
```

# 📋 Referência Rápida

| Biblioteca | Comando | O que faz |
|------------|---------|----------|
| `np` | `np.array([...])` | Cria array |
| `np` | `np.mean() / std() / sum()` | Estatísticas |
| `np` | `np.zeros() / ones() / arange()` | Arrays prontos |
| `np` | `np.argmax() / argmin()` | Índice do maior/menor |
| `np` | `array[array > valor]` | Filtragem booleana |
| `pd` | `pd.read_csv('arquivo.csv')` | Lê CSV |
| `pd` | `df.head() / tail() / info()` | Inspecionar |
| `pd` | `df[condição]` | Filtrar linhas |
| `pd` | `df.groupby().agg()` | Agrupar |
| `pd` | `df['col'].apply(func)` | Aplicar função |
| `pd` | `df.to_csv('saida.csv', index=False)` | Salvar CSV |
| `plt` | `plt.plot() / bar() / scatter() / hist()` | Gráficos |
| `plt` | `plt.subplots(n, m)` | Múltiplos gráficos |
| `plt` | `plt.savefig('img.png', dpi=150)` | Salvar imagem |
