# PARTE 1 — Condicionais (`if / elif / else`)

Um condicional permite que o programa execute ações diferentes dependendo de uma condição.

```python
if condição:
    # faz isso se a condição for verdadeira
elif outra_condição:
    # faz isso se a segunda condição for verdadeira
else:
    # faz isso se nenhuma condição for verdadeira
```

Os operadores de comparação mais comuns são:

| Operador | Significado |
|---|---|
| `==` | igual a |
| `!=` | diferente de |
| `>` | maior que |
| `<` | menor que |
| `>=` | maior ou igual |
| `<=` | menor ou igual |

## Exemplo 1.1 — Classificação pelo NSPT

O número de golpes do SPT (NSPT) indica a consistência/compacidade do solo. Veja a classificação usual para argilas conforme a NBR 6484/2020:

| NSPT | Consistência |
|---|---|
| ≤ 2 | Muito mole |
| 3 – 5 | Mole |
| 6 – 10 | Média |
| 11 – 19 | Rija |
| 20 – 30 | Muito rija |
| > 30 | Dura |



````python
#Argilas
Nspt = 8

if Nspt <= 2:
    consistencia = "Muito mole"
elif Nspt <= 5:
    consistencia = "Mole"
elif Nspt <= 10:
    consistencia = "Média"
elif Nspt <= 19:
    consistencia = "Rija"
elif Nspt <= 30:
    consistencia = "Muito rija"
else:
    consistencia = "Dura"

print("NSPT:", Nspt)
print("Consistência da argila:", consistencia)
````

## Exemplo 1.3 — Verificação do nível d'água

Verificamos se um ponto de análise está acima ou abaixo do nível d'água (NA).

````python
profundidade_NA = 3.0   # Nível d'água (m)
z = 5.0                 # Profundidade do ponto de análise (m)

if z < profundidade_NA:
    print("Ponto acima do NA — solo não saturado.")
    print("Usar peso específico natural.")
elif z == profundidade_NA:
    print("Ponto exatamente no NA.")
else:
    print("Ponto abaixo do NA — solo saturado.")
    print("Considerar pressão de poros no cálculo.")
````

# PARTE 2 — Repetições (`for` e `while`)

Repetições permitem executar um bloco de código várias vezes automaticamente — perfeito para processar várias camadas de solo, vários furos de sondagem, etc.

## 2.1 — O laço `for`

O `for` percorre uma sequência de valores:

```python
for variavel in sequencia:
    # repete para cada valor
```

## Exemplo 2.1 — Tensão vertical em várias profundidades

Calcular σᵥ a cada metro, de 1 até 6 m.

````python
gamma = 18.0   # kN/m³

print("Profundidade (m) | Tensão vertical (kPa)")
print("-" * 40)

for z in range(1, 7):          # de 1 até 6 metros
    sigma_v = gamma * z
    print(f"       {z} m        |       {sigma_v:.1f} kPa")
````

> 💡 `range(1, 7)` gera os números 1, 2, 3, 4, 5, 6 — o último valor não é incluído.
>
> `f"...{variavel:.1f}..."` é uma forma prática de formatar números dentro de textos — o `.1f` significa 1 casa decimal.

## Exemplo 2.2 — Classificar uma lista de valores NSPT

Temos os resultados de SPT de várias camadas e queremos classificar todas de uma vez.

````python
# Resultados de NSPT por camada (metro a metro)
nspt_camadas = [2, 4, 7, 12, 18, 25, 30]

print("Camada | NSPT | Consistência")
print("-" * 35)

for i, Nspt in enumerate(nspt_camadas):
    if Nspt <= 2:
        consistencia = "Muito mole"
    elif Nspt <= 5:
        consistencia = "Mole"
    elif Nspt <= 10:
        consistencia = "Média"
    elif Nspt <= 19:
        consistencia = "Rija"
    elif Nspt <= 30:
        consistencia = "Muito rija"
    else:
        consistencia = "Dura"

    print(f"  {i+1}    |  {Nspt:2d}  | {consistencia}")
````

> 💡 `enumerate()` retorna o índice e o valor ao mesmo tempo — muito útil para numerar camadas.

## 2.2 — O laço `while`

O `while` repete enquanto uma condição for verdadeira:

```python
while condição:
    # repete enquanto a condição for verdadeira
```

## Exemplo 2.3 — Aprofundamento até atingir camada resistente

Simula a sondagem avançando metro a metro até encontrar NSPT ≥ 20 (camada dura).

````python
# NSPT medido a cada metro de sondagem
sondagem = [3, 5, 8, 11, 15, 22, 28]

profundidade = 1
camada = 0

print("Avançando a sondagem...")
print("-" * 30)

while camada < len(sondagem):
    Nspt = sondagem[camada]
    print(f"Profundidade {profundidade} m → NSPT = {Nspt}")

    if Nspt >= 20:
        print(f"\n✅ Camada resistente encontrada a {profundidade} m!")
        break   # para o laço

    profundidade += 1
    camada += 1
````

> 💡 `break` interrompe o laço imediatamente quando a condição desejada é atingida.
